#Feature Engineering - Handling Imabalanced Dataset 
--Performing upsampling (Increase minority datapoints) and 
--Downsampling (Reduce majority datapoints)

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(123)

# Create a sample imbalanced dataset
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# Creating my DataFrame with imbalace dataset
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=5, scale=2, size=n_class_0),
    'target':[0]*n_class_0
})
class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature2': np.random.normal(loc=7, scale=2, size=n_class_1),
    'target':[1]*n_class_1
})

# Combine the two classes to form the imbalanced dataset

df = pd.concat([class_0, class_1]).reset_index(drop=True)
print(df.head())
print(df.tail())
```

```python
df['target'].value_counts()
```

```python
#Upsampling
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, replace=True,
                                n_samples=len(df_majority),
                                random_state=123)
df_minority_upsampled.head()
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
print(df_upsampled.shape)
print(df_upsampled['target'].value_counts())
```

```python
# Downsampling
df_majority_downsampled = resample(df_majority, replace=False,
                                  n_samples=len(df_minority),
                                  random_state=123)
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
print(df_downsampled.shape)
print(df_downsampled['target'].value_counts())
```
